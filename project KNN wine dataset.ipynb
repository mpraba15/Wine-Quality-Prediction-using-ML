```python
#KNN for Classification Using scikit-learn
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Wine dataset
datas = load_wine()
```

```python
data = pd.DataFrame(datas.data, columns=datas.feature_names)
data
```

```python
x = datas.data
y = datas.target
```

```python
datas.target
```

```python
# Split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# Scale the features
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Train the KNN model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
```

```python
# Make predictions
y_pred = knn.predict(x_test)
y_pred
```

```python
# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print("Classification Report:\n", classification_report(y_test, y_pred))
```

#Hyperparameter Tuning

```python
# Define the KNN model
knn = KNeighborsClassifier()
```

```python
# Define the hyperparameter grid
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],  # Values for k
    'metric': ['euclidean', 'manhattan', 'minkowski']  # Distance metrics
}
```

```python
# Use GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, cv=5, scoring='accuracy', verbose=1)
```

```python
# Fit the model
grid_search.fit(x_train, y_train)
```

```python
# Best hyperparameters
print("Best hyperparameters:", grid_search.best_params_)
```

```python
# Evaluate the best model on the test set
best_knn = grid_search.best_estimator_
y_pred = best_knn.predict(x_test)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
```
